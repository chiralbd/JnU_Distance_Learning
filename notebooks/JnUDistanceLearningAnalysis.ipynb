## Mental Disturbance

```python
rp.summary_cat(data['did_you_feel_any_mental_disturbance_during_online_class?'])
```

```python
# https://stackoverflow.com/questions/20162926/process-multiple-answer-questionnaire-from-google-forms-results-with-pandas
data['what_kind_of_mental_disturbance_did_you_feel_during_online_class?']
```

```python
data['what_kind_of_mental_disturbance_did_you_feel_during_online_class?'].value_counts()
```

```python
mental_problems = data['what_kind_of_mental_disturbance_did_you_feel_during_online_class?'].str.get_dummies(sep=', ')
```

```python
mental_problems
```

```python
mental_problems = mental_problems.replace({0: "No", 1: 'Yes'})
```

```python
mental_problems.columns = mental_problems.columns.str.strip() 

mental_problems.columns
```

```python
rp.summary_cat(mental_problems)
```
